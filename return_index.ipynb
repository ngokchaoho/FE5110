```python
import pandas as pd
pd.options.display.max_rows = 4000
```

```python
curr_month = pd.read_excel(io=r"C:\Users\HO NGOK CHAO\PycharmProjects\FE5110\Data\SGX Baltic Capesize Time Charter Basket (4 Routes) FFA0.xls",
                           header=None,
                           names=['date','close_prcie'],
                           skiprows=9,
                           skipfooter=2,
                           )
curr_month.set_index('date',inplace=True)
date_index = pd.date_range(start=min(curr_month.index),end=max(curr_month.index))
curr_month.reindex(date_index, method='ffill')
curr_month['return_rate'] = curr_month.shift(-1) / curr_month  - 1
```

```python
next_month = pd.read_excel(io=r"C:\Users\HO NGOK CHAO\PycharmProjects\FE5110\Data\SGX Baltic Capesize Time Charter Basket (4 Routes) FFA1.xls",
                           header=None,
                           names=['date','close_prcie'],
                           skiprows=9,
                           skipfooter=2,
                           )
next_month.set_index('date',inplace=True)
date_index = pd.date_range(start=min(next_month.index),end=max(next_month.index))
next_month.reindex(date_index, method='ffill')
next_month['return_rate'] = next_month.shift(-1) / next_month  - 1
```

```python
rolling_date = 15
curr_month = curr_month.loc[curr_month.index.day<rolling_date,:]
next_month = next_month.loc[next_month.index.day>=rolling_date,:]
```

```python
return_series = pd.concat([curr_month,next_month])
return_series.sort_index(inplace=True)
```
